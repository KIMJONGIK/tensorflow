# merge

```python
df1 = pd.DataFrame({
    '고객번호' : [1001, 1002, 1002, 1003, 1004, 1005],
    '이름'     : ['둘리', '도우너', '또치', '고길동', '마이콜', '홍영희']
}, columns=['고객번호', '이름'])
df1.head()
```

```python
df2 = pd.DataFrame({
    '고객번호' : [1001, 1001, 1002, 1003, 1004],
    '금액'     : [10000, 3000, 20000, 15000, 30000]
}, columns=['고객번호', '금액'])
df2.head()
```

```python
df = pd.merge(df1, df2)
df
```

```python
df = pd.merge(df1, df2, how='outer')
df
```

```python
# outer와 같은 결과
df = pd.merge(df1, df2, how='left')
df
```

```python
# inner와 같은 결과
df = pd.merge(df1, df2, how='right')
df
```

```python
# on은 양쪽에 있는 데이터중 어떤 것을 기준으로 할 것인가
df = pd.merge(df1, df2, on='고객번호')
df
```

```python
df = pd.merge(df1, df2,left_on='고객번호', right_on='고객번호')
df
```

## 인덱스 기준

```python
df1 = pd.DataFrame({
    '도시' : ['서울', '서울', '서울', '서울', '부산', '부산'],
    '연도' : [2000, 2003, 2009, 2010, 2005, 2006],
    '인구' : [9835959, 9943234, 10000034,10000054, 5000000, 6000000 ]   
})
df1
```

```python
df2 = pd.DataFrame(
    np.arange(12).reshape((-1,2)),
    index=[['부산', '부산', '서울', '서울', '서울', '서울'],
          [2005, 2006, 2000, 2003, 2009, 2010]],
    columns=['데이터1', '데이터2']
)
df2
```

```python
df_r = pd.merge(df1, df2, left_on=['도시', '연도'], right_index=True)
df_r
```

```python
df = pd.read_excel('02-sales-funnel.xlsx')
df.head(3)
```

```python
# values를 넣으면 원하는 컬럼명만 볼 수 있다.
pd.pivot_table(df, index=['Name', 'Rep', 'Manager'], values=['Price'])
```

```python
# aggfunc에 함수를 넣을 수 있다.
pd.pivot_table(df, index=['Name', 'Rep', 'Manager'],
               values=['Price'],
              aggfunc=[np.sum, np.mean, np.count_nonzero])
```

```python
# columns를 지정하면 원하는 항목별로 볼 수 있다.
pd.pivot_table(df, index=['Rep', 'Manager'],
               values=['Price'],
              columns=['Product'],
              aggfunc=[np.sum, np.mean, np.count_nonzero])
```
